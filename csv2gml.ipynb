```python
import pandas as pd
import networkx as nx
```

```python
# Load the CSV file
csv_file = "/content/Emergency Team(Edges).csv"  # Change to your file name
df = pd.read_csv(csv_file)
```

# CSV to GML

```python

# Create an empty graph
G = nx.Graph()

# Process the edges
for _, row in df.iterrows():
    source = row['Source']
    targets = str(row['Target']).split(',')  # Split multiple targets
    for target in targets:
        G.add_edge(str(source), target.strip())  # Ensure nodes are strings for consistency

# Save to GML format
gml_file = "rehabilitation_team_network.gml"
nx.write_gml(G, gml_file)

print(f"GML file saved as: {gml_file}")
```

```python
# Print GML Data
with open(gml_file, "r") as f:
    gml_data = f.read()
    print(gml_data)  # Prints the GML content to console

print(f"GML file saved as: {gml_file}")
```

# CSV to CSV only edge separating


```python

# Prepare a list to store separated edges
edges = []

# Process the edges
for _, row in df.iterrows():
    source = row['Source']
    targets = str(row['Target']).split(',')  # Split multiple targets
    for target in targets:
        edges.append([source, target.strip()])  # Append clean edges to the list

# Convert list to DataFrame
edges_df = pd.DataFrame(edges, columns=["Source", "Target"])

# Save to CSV
output_csv = "rehabilitation_team_edges_cleaned.csv"
edges_df.to_csv(output_csv, index=False)

print(f"Cleaned CSV file saved as: {output_csv}")
```
